```python
data4 = {
    'emp_id': [101, 102, 103, 104], 
    'salary': [234, 345, 200, 400], 
    'pf':[23, 34, 56, 78]
}

data5 = { 
    'emp_id': [101, 102, 103, 105], 
    'mob_no': [3435, 5646, 12312, 12312], 
    'house_nu': [345, 645, 213, 234]
}
```

```python
df4 = pd.DataFrame(data4)
df4
```

```python
df5 = pd.DataFrame(data5)
df5
```

```python
# Merge Operation 
pd.merge(df4, df5)
```

```python
pd.merge(df4, df5, how='left') # Left Merge
```

```python
pd.merge(df4, df5, how='right') # Right Merge
```

```python
data6 = {
    'emp_id1': [101, 102, 103, 104], 
    'salary': [234, 345, 200, 400], 
    'pf':[23, 34, 56, 78]
}

data7 = {
    'emp_id2': [101, 102, 103, 105], 
    'mob_no': [3435, 5646, 12312, 12312], 
    'house_nu': [345, 645, 213, 234]
}

df6 = pd.DataFrame(data6)
df7 = pd.DataFrame(data7)
```

```python
pd.merge(df6, df7, left_on='emp_id1', right_on='emp_id2')
```
